```python
import pandas as pd
import numpy as np
import re
```

```python
jails_df = pd.read_csv("all_jails.csv")
```

```python
df= jails_df.dropna(subset=['med2008', 'med2009', 'med2010', 'med2011', 'med2012', 'med2013', 'med2014', 'med2015', 'med2016', 'med2017', 'med2018', 'med2019'])
```

```python
df.loc[df['med2008'].str.contains('/'), 'med2008'] = 'transition'
df.loc[df['med2009'].str.contains('/'), 'med2009'] = 'transition'
df.loc[df['med2010'].str.contains('/'), 'med2010'] = 'transition'
df.loc[df['med2011'].str.contains('/'), 'med2011'] = 'transition'
df.loc[df['med2012'].str.contains('/'), 'med2012'] = 'transition'
df.loc[df['med2013'].str.contains('/'), 'med2013'] = 'transition'
df.loc[df['med2014'].str.contains('/'), 'med2014'] = 'transition'
df.loc[df['med2015'].str.contains('/'), 'med2015'] = 'transition'
df.loc[df['med2016'].str.contains('/'), 'med2016'] = 'transition'
df.loc[df['med2017'].str.contains('/'), 'med2017'] = 'transition'
df.loc[df['med2018'].str.contains('/'), 'med2018'] = 'transition'
df.loc[df['med2019'].str.contains('/'), 'med2019'] = 'transition'

```

```python
df['med2008'] = np.where(~df['med2008'].str.contains("public|transition"), "private", df['med2008'])
df['med2009'] = np.where(~df['med2009'].str.contains("public|transition"), "private", df['med2009'])
df['med2010'] = np.where(~df['med2010'].str.contains("public|transition"), "private", df['med2010'])
df['med2011'] = np.where(~df['med2011'].str.contains("public|transition"), "private", df['med2011'])
df['med2012'] = np.where(~df['med2012'].str.contains("public|transition"), "private", df['med2012'])
df['med2013'] = np.where(~df['med2013'].str.contains("public|transition"), "private", df['med2013'])
df['med2014'] = np.where(~df['med2014'].str.contains("public|transition"), "private", df['med2014'])
df['med2015'] = np.where(~df['med2015'].str.contains("public|transition"), "private", df['med2015'])
df['med2016'] = np.where(~df['med2016'].str.contains("public|transition"), "private", df['med2016'])
df['med2017'] = np.where(~df['med2017'].str.contains("public|transition"), "private", df['med2017'])
df['med2018'] = np.where(~df['med2018'].str.contains("public|transition"), "private", df['med2018'])
df['med2019'] = np.where(~df['med2019'].str.contains("public|transition"), "private", df['med2019'])


```

```python
df['death_proportion2008'] = df.d2008 / df.adp2008
df['death_proportion2009'] = df.d2009 / df.adp2009
df['death_proportion2010'] = df.d2010 / df.adp2010
df['death_proportion2011'] = df.d2011 / df.adp2011
df['death_proportion2012'] = df.d2012 / df.adp2012
df['death_proportion2013'] = df.d2013 / df.adp2013
df['death_proportion2014'] = df.d2014 / df.adp2014
df['death_proportion2015'] = df.d2015 / df.adp2015
df['death_proportion2016'] = df.d2016 / df.adp2016
df['death_proportion2017'] = df.d2017 / df.adp2017
df['death_proportion2018'] = df.d2018 / df.adp2018
df['death_proportion2019'] = df.d2019 / df.adp2019

```

```python
df2008 = df[['med2008', 'death_proportion2008']]
df2009 = df[['med2009', 'death_proportion2009']]
df2010 = df[['med2010', 'death_proportion2010']]
df2011 = df[['med2011', 'death_proportion2011']]
df2012 = df[['med2012', 'death_proportion2012']]
df2013 = df[['med2013', 'death_proportion2013']]
df2014 = df[['med2014', 'death_proportion2014']]
df2015 = df[['med2015', 'death_proportion2015']]
df2016 = df[['med2016', 'death_proportion2016']]
df2017 = df[['med2017', 'death_proportion2017']]
df2018 = df[['med2018', 'death_proportion2018']]
df2019 = df[['med2019', 'death_proportion2019']]

```

```python
df_2008 = df2008.rename(columns={"med2008": "med", "death_proportion2008": "death_proportion"})
df_2009 = df2009.rename(columns={"med2009": "med", "death_proportion2009": "death_proportion"})
df_2010 = df2010.rename(columns={"med2010": "med", "death_proportion2010": "death_proportion"})
df_2011 = df2011.rename(columns={"med2011": "med", "death_proportion2011": "death_proportion"})
df_2012 = df2012.rename(columns={"med2012": "med", "death_proportion2012": "death_proportion"})
df_2013 = df2013.rename(columns={"med2013": "med", "death_proportion2013": "death_proportion"})
df_2014 = df2014.rename(columns={"med2014": "med", "death_proportion2014": "death_proportion"})
df_2015 = df2015.rename(columns={"med2015": "med", "death_proportion2015": "death_proportion"})
df_2016 = df2016.rename(columns={"med2016": "med", "death_proportion2016": "death_proportion"})
df_2017 = df2017.rename(columns={"med2017": "med", "death_proportion2017": "death_proportion"})
df_2018 = df2018.rename(columns={"med2018": "med", "death_proportion2018": "death_proportion"})
df_2019 = df2019.rename(columns={"med2019": "med", "death_proportion2019": "death_proportion"})
```

```python
final_df = pd.concat([df_2008, df_2009, df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019], axis=0)
```
